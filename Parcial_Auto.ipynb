```python
try:
    año=int(input("Ingrese el año del que desea saber cuántas canciones se lanzaron antes de ese año: "))
    def canciones_antes_de(df, año):
        return df[df["released_year"] < año]["track_name"].tolist()

    ki=canciones_antes_de(df, año)
    print(ki)
    print("Número de canciones:",  len(ki))
except ValueError:
    print("Ingrese un parámetro correcto")
```

```python
#Número de canciones lanzadas en un año
try:
    hola=int(input("Ingrese el año que desea saber el número de canciones lanzadas desde 1930 hasta el 2023: "))
    lanzadas = df[df["released_year"] == hola]
    print("Número de canciones lanzadas en ese año:", len(lanzadas))
except ValueError:
    print("Ingrese un parámetro correcto")
```

```python
#Cree una tabla con una función de agregación, que cuente cuantas canciones hay de Taylor Swift y cuantas hay de Adele en la base de datos.
import numpy as np
import matplotlib.pyplot as plt

taylor = df[df["artist(s)_name"] == "Taylor Swift"]
adele = df[df["artist(s)_name"] == "Adele"]
print("Número de canciones de Taylor Swift:", len(taylor))
print("Número de canciones de Adele:", len(adele))

tabla = df[df["artist(s)_name"].isin(["Taylor Swift", "Adele"])]
print(tabla)
```
